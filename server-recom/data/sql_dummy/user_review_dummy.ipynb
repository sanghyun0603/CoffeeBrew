```python
# sql_bean.csv
bean_read = pd.read_csv(path.join('.', 'sql_bean.csv'), low_memory=False, encoding='utf-8')

print(bean_read.shape)
bean_read.head()
```

```python
# sql_capsule.csv
capsule_read = pd.read_csv(path.join('.', 'sql_capsule.csv'), low_memory=False, encoding='utf-8')

print(capsule_read.shape)
capsule_read.head()
```

```python
# sql_capsule_score.csv
capsule_score_read = pd.read_csv(path.join('.', 'sql_capsule_score.csv'), low_memory=False, encoding='utf-8')

print(capsule_score_read.shape)
capsule_score_read.head()
```

```python
capsule_data = pd.merge(capsule_read, capsule_score_read, left_on='idx', right_on='capsule_idx')
print(capsule_data.shape)
capsule_data.head()
```

```python
capsule_data.rename({'idx_x':'idx'}, axis=1, inplace=True)
capsule_data.drop(['created_date_x', 'updated_date_x', 'name_en', 'summary', 'thumbnail', 'user_grade', 'idx_y', 'created_date_y', 'updated_date_y', 'capsule_idx'], axis=1, inplace=True)
print(capsule_data.shape)
capsule_data.head()
```

```python
# sql_member.csv
member_read = pd.read_csv(path.join('.', 'sql_member.csv'), low_memory=False, encoding='utf-8')

print(member_read.shape)
member_read.head()
```

```python
def generate_random_overall(item_len):
    def random_overall():
        return random.choices([i for i in range(1, 11)], weights=[4, 4, 3, 3, 5, 5, 7, 10, 10, 8])
    
    random_list = [random.choices([random_overall()[0], np.NaN], weights=[item_len * 0.15, item_len * 0.85])[0] for i in range(item_len)]
    random_series = pd.Series(random_list, index=[i+1 for i in range(item_len)])
    
    return random_series
```

```python
generate_random_overall(20)
```

```python
overall_bean_df = pd.DataFrame(index=member_read['idx'], columns=bean_data['idx'])
overall_bean_df.index.rename('member_idx', inplace=True)
overall_bean_df.columns.rename('bean_idx', inplace=True)
print(overall_bean_df.shape)
overall_bean_df.head()
```

```python
overall_bean_df = overall_bean_df.apply(lambda x: generate_random_overall(overall_bean_df.shape[1]), axis=1)
overall_bean_df.columns.rename('item_idx', inplace=True)
overall_bean_df.head()
```

```python
max_idx = overall_bean_df.notnull().sum(axis=1).idxmax()
min_idx = overall_bean_df.notnull().sum(axis=1).idxmin()

print('max: ', max_idx, overall_bean_df.loc[max_idx].notnull().sum().max())
print('min: ', min_idx, overall_bean_df.loc[min_idx].notnull().sum().min())
```

```python
overall_bean_df = overall_bean_df.melt(ignore_index=False).reset_index()
overall_bean_df.dropna(inplace=True)
overall_bean_df.reset_index(drop=True, inplace=True)
overall_bean_df.rename({'value':'overall'}, axis=1, inplace=True)
overall_bean_df = overall_bean_df.astype({'overall':'int'})
print(overall_bean_df.shape)
overall_bean_df.head()
```

```python
review_bean_df = pd.merge(overall_bean_df, bean_data.drop(['name_ko', 'roasting_point'], axis=1), how='left', left_on='item_idx', right_on='idx').drop(['idx', 'balance'], axis=1)
print(review_bean_df.shape)
review_bean_df.head()
```

```python
review_bean_df['idx'] = review_bean_df.index + 1
review_bean_df['created_date'] = member_read['created_date'].iloc[0]
review_bean_df['updated_date'] = member_read['created_date'].iloc[0]
review_bean_df['expired'] = 'N'
review_bean_df['content'] = '커피가 괜찮아요'
review_bean_df['item_type'] = 'bean'
review_bean_df['like'] = 0
overall_list_columns = ['idx', 'created_date', 'updated_date', 'acidity', 
                        'bitterness', 'body', 'coffeeing_note', 'content', 
                        'expired', 'flavor', 'item_idx', 'item_type', 
                        'like', 'overall', 'sweetness', 'member_idx']
review_bean_df = review_bean_df[overall_list_columns]
print(review_bean_df.shape)
review_bean_df.head()
```

```python
overall_capsule_df = pd.DataFrame(index=member_read['idx'], columns=capsule_data['idx'])
overall_capsule_df.index.rename('member_idx', inplace=True)
overall_capsule_df.columns.rename('capsule_idx', inplace=True)
print(overall_capsule_df.shape)
overall_capsule_df.head()
```

```python
overall_capsule_df = overall_capsule_df.apply(lambda x: generate_random_overall(overall_capsule_df.shape[1]), axis=1)
overall_capsule_df.columns.rename('item_idx', inplace=True)
overall_capsule_df.head()
```

```python
max_idx = overall_capsule_df.notnull().sum(axis=1).idxmax()
min_idx = overall_capsule_df.notnull().sum(axis=1).idxmin()

print('max: ', max_idx, overall_capsule_df.loc[max_idx].notnull().sum().max())
print('min: ', min_idx, overall_capsule_df.loc[min_idx].notnull().sum().min())
```

```python
overall_capsule_df = overall_capsule_df.melt(ignore_index=False).reset_index()
overall_capsule_df.dropna(inplace=True)
overall_capsule_df.reset_index(drop=True, inplace=True)
overall_capsule_df.rename({'value':'overall'}, axis=1, inplace=True)
overall_capsule_df = overall_capsule_df.astype({'overall':'int'})
print(overall_capsule_df.shape)
overall_capsule_df.head()
```

```python
review_capsule_df = pd.merge(overall_capsule_df, capsule_data.drop(['name_ko'], axis=1), how='left', left_on='item_idx', right_on='idx').drop(['idx', 'balance'], axis=1)
print(review_capsule_df.shape)
review_capsule_df.head()
```

```python
review_capsule_df['idx'] = review_capsule_df.index + review_bean_df.shape[0] + 1
review_capsule_df['created_date'] = member_read['created_date'].iloc[0]
review_capsule_df['updated_date'] = member_read['created_date'].iloc[0]
review_capsule_df['expired'] = 'N'
review_capsule_df['content'] = '커피가 괜찮아요'
review_capsule_df['item_type'] = 'capsule'
review_capsule_df['like'] = 0

review_capsule_df['sweetness'] = review_capsule_df['roasting']

overall_list_columns = ['idx', 'created_date', 'updated_date', 'acidity', 
                        'bitterness', 'body', 'coffeeing_note', 'content', 
                        'expired', 'flavor', 'item_idx', 'item_type', 
                        'like', 'overall', 'sweetness', 'member_idx']
review_capsule_df = review_capsule_df[overall_list_columns]
print(review_capsule_df.shape)
review_capsule_df.head()
```

```python
review_df = pd.concat([review_bean_df, review_capsule_df], ignore_index=True)
review_df
```

```python
# 파일 저장
os.makedirs(DIR_SAVE_PATH, exist_ok=True)
review_df.to_csv(path.join(DIR_SAVE_PATH, 'sql_review_new.csv'), sep=',', index=False, encoding='utf-8')
```

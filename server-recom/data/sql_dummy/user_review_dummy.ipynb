```python
# sql_bean.csv
bean_read = pd.read_csv(path.join('.', 'sql_bean.csv'), low_memory=False, encoding='cp949')

print(bean_read.shape)
bean_read.head()
```

```python
# sql_bean_score.csv
bean_score_read = pd.read_csv(path.join('.', 'sql_bean_score.csv'), low_memory=False, encoding='cp949')

print(bean_score_read.shape)
bean_score_read.head()
```

```python
bean_data = pd.merge(bean_read, bean_score_read, left_on='idx', right_on='bean_idx')
print(bean_data.shape)
bean_data.head()
```

```python
bean_data.rename({'idx_x':'idx'}, axis=1, inplace=True)
bean_data.drop(['created_date_x', 'updated_date_x', 'name_en', 'summary', 'thumbnail', 'user_grade', 'idx_y', 'created_date_y', 'updated_date_y', 'bean_idx'], axis=1, inplace=True)
print(bean_data.shape)
bean_data.head()
```

```python
# sql_member.csv
user_read = pd.read_csv(path.join('.', 'sql_member.csv'), low_memory=False, encoding='cp949')

print(user_read.shape)
user_read.head()
```

```python
def generate_random_overall(item_len):
    def random_overall():
        return random.choices([i for i in range(1, 11)], weights=[4, 4, 3, 3, 5, 5, 7, 10, 10, 8])
    
    random_list = [random.choices([random_overall()[0], np.NaN], weights=[item_len * 0.2, item_len * 0.8])[0] for i in range(item_len)]
    random_series = pd.Series(random_list, index=[i+1 for i in range(item_len)])
    
    return random_series
```

```python
generate_random_overall(20)
```

```python
overall_list_df = pd.DataFrame(index=user_read['idx'], columns=bean_data['idx'])
overall_list_df.index.rename('member_idx', inplace=True)
overall_list_df.columns.rename('bean_idx', inplace=True)
print(overall_list_df.shape)
overall_list_df.head()
```

```python
overall_list_df = overall_list_df.apply(lambda x: generate_random_overall(overall_list_df.shape[1]), axis=1)
overall_list_df.columns.rename('item_idx', inplace=True)
```

```python
overall_list_df.head()
```

```python
max_idx = overall_list_df.notnull().sum(axis=1).idxmax()
min_idx = overall_list_df.notnull().sum(axis=1).idxmin()

print('max: ', max_idx, overall_list_df.loc[max_idx].notnull().sum().max())
print('min: ', min_idx, overall_list_df.loc[min_idx].notnull().sum().min())
```

```python
overall_list_df = overall_list_df.melt(ignore_index=False).reset_index()
overall_list_df.dropna(inplace=True)
overall_list_df.reset_index(drop=True, inplace=True)
overall_list_df.rename({'value':'overall'}, axis=1, inplace=True)
overall_list_df = overall_list_df.astype({'overall':'int'})
print(overall_list_df.shape)
overall_list_df.head()
```

```python
review_df = pd.merge(overall_list_df, bean_data.drop(['name_ko', 'roasting_point'], axis=1), how='left', left_on='item_idx', right_on='idx').drop(['idx', 'balance'], axis=1)
print(review_df.shape)
review_df.head()
```

```python
review_df['idx'] = review_df.index + 1
review_df['created_date'] = ''
review_df['updated_date'] = ''
review_df['expired'] = 'N'
review_df['content'] = '커피가 괜찮아요'
review_df['item_type'] = 'bean'
review_df['like'] = 0
overall_list_columns = ['idx', 'created_date', 'updated_date', 'acidity', 
                        'bitterness', 'body', 'coffeeing_note', 'content', 
                        'expired', 'flavor', 'item_idx', 'item_type', 
                        'like', 'overall', 'sweetness', 'member_idx']
review_df = review_df[overall_list_columns]
print(review_df.shape)
review_df.head()
```

```python
# 파일 저장
os.makedirs(DIR_SAVE_PATH, exist_ok=True)
review_df.to_csv(path.join(DIR_SAVE_PATH, 'sql_review_new.csv'), sep=',', index=False, encoding='cp949')
```

```python
# 고도 단위에서 결측치를 확인하고, 결측치를 공백(' ')으로 채움
print('na: {}'.format(df_arabica['altitude_unit'].isna().sum()))

df_arabica['altitude_unit'].fillna(' ', inplace=True)
print(df_arabica['altitude_unit'].value_counts(dropna=False))
```

```python
# 고도 정보의 결측치를 확인하고, 결측치가 있으면 na로 채움
print('na: {}'.format(df_arabica['altitude'].isna().sum()))

df_arabica['altitude'].fillna('na', inplace=True)
print(df_arabica['altitude'].value_counts(dropna=False))
```

```python
# 고도와 고도 단위를 결합해서 하나의 문자열로 만듬
df_altitude = (df_arabica['altitude'] + ' ' + df_arabica['altitude_unit']).str.strip()
print(df_altitude)
```

```python
# 기존의 altitude column을 결합치로 치환
df_arabica['altitude'] = df_altitude
print(df_arabica['altitude'])
```

```python
# 고도 정보가 있으면 숫자가 하나라도 포함하고 있을 것이라 판단
# str.contains(r'\d')을 통해서 숫자를 하나도 포함하지 않는 행 정보를 스캔
arabica_altitude_na = ~df_arabica['altitude'].str.contains(r'\d')
arabica_altitude_na
```

```python
# 찾아낸 이상 값을 결측치를 -1로 채움 (고도 0인 값과 구분하기 위함)
# 상품으로 판매되는 모든 커피는 해발고도 0m보다 높은 곳에서 생산된다는 정보를 전제로 함 
df_arabica.loc[arabica_altitude_na, 'altitude'] = '-1'
print(df_arabica.loc[df_arabica['altitude'] == '-1'][['random_id', 'altitude']])
```

```python
# 유의미한 데이터를 담는 것으로 추정되는 columns를 list로 정의
df_arabica_use_col = ['random_id', 'species', 'origin', 'region', 'country', 'farm', 'altitude', 'grade', 'green_pass', 'isCertified']
```

```python
# 데이터 출력
df_arabica[df_arabica_use_col]
```
